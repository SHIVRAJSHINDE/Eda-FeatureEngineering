```python
import pandas as pd

file_path = "D://MLProjects//ZDatasets//NLP//SPAM Classification//spam.csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')
df.head()
```

```python
df.groupby('v1').describe()
```

```python

df['v1'].value_counts()
```

```python
hamDf = df[df['v1']=='ham']
hamDf
```

```python
spamDf = df[df['v1']=='spam']
spamDf
spamDf.shape[0]
```

```python
hamDf = hamDf.sample(spamDf.shape[0],random_state=42)
hamDf
```

```python
df_Balanced = pd.concat([hamDf,spamDf])
df_Balanced
```

```python
df_Balanced['v1'].value_counts()
```

```python
df_Balanced['spam'] = df_Balanced['v1'].apply(lambda x : 1 if x=="spam" else 0)

df_Balanced
```
