```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df['ProdTaken'], test_size=0.33, random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder

```

```python
li2 = ['ProductPitched','Designation']
li2[0]
```

```python
transfor = ColumnTransformer(transformers=[
    ('trans1',OrdinalEncoder(categories=[['Manager', 'Executive', 'Senior Manager', 'AVP', 'VP']]),['Designation']),
    ('trans2',OrdinalEncoder(categories=[['Basic','Standard','Deluxe','Super Deluxe', 'King']]),['ProductPitched']),
    ('trans3',OneHotEncoder(sparse=False,drop='first'),['TypeofContact','Occupation','Gender','MaritalStatus'])
    ],remainder='passthrough')
```

```python
X_train = transfor.fit_transform(X_train)
X_test = transfor.transform(X_test)
```

```python
from sklearn.svm import SVC

modelSVC = SVC()
```

```python
modelSVC.fit(X_train,y_train)
```

```python
modelSVC.score(X_train,y_train)
```

```python
y_predict = modelSVC.predict(X_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```
