```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python

df = pd.read_excel('tourismData.xlsx','Tourism')
df
```

```python
df = df[['ProdTaken', 'Age', 'TypeofContact', 'CityTier', 'Occupation', 'Gender', 'NumberOfPersonVisited',
       'NumberOfFollowups', 'ProductPitched', 'PreferredPropertyStar',
       'MaritalStatus', 'NumberOfTrips', 'Passport', 'FeedbackScore', 'OwnCar',
       'NumberOfChildrenVisited', 'Designation', 'MonthlyIncome']]

df
```

```python
df.info()
```

```python
# Calculate the total missing values in each column
missing_values = df.isnull().sum()

missing_values
```

```python
# Filter columns with missing values greater than 0
columns_with_missing = missing_values[missing_values > 0]
columns_with_missing

```

```python
# Print the column names with missing values
for column_name, missing_count in missing_values.items():
    if missing_count > 0:
        print(column_name, missing_count)
```

```python
AgeMedian = df['Age'].median()
NumberOfFollowupsMedian = df['NumberOfFollowups'].median()
TypeofContactMode = df['TypeofContact'].mode()[0]
PreferredPropertyStarMode = df['PreferredPropertyStar'].mode()[0]
NumberOfTripsMedian = df['NumberOfTrips'].median()
NumberOfChildrenVisitedMedian = df['NumberOfChildrenVisited'].median()
MonthlyIncomeMedian = df['MonthlyIncome'].median()
```

```python
df['Age'].fillna(AgeMedian,inplace=True)
df['TypeofContact'].fillna(TypeofContactMode,inplace=True)
df['NumberOfFollowups'].fillna(NumberOfFollowupsMedian,inplace=True)
df['PreferredPropertyStar'].fillna(PreferredPropertyStarMode,inplace=True)
df['NumberOfTrips'].fillna(NumberOfTripsMedian,inplace=True)
df['NumberOfChildrenVisited'].fillna(NumberOfChildrenVisitedMedian,inplace=True)
df['MonthlyIncome'].fillna(MonthlyIncomeMedian,inplace=True)

missing_values = df.isnull().sum()

# Print the column names with missing values
for column_name, missing_count in missing_values.items():
    if missing_count > 0:
        print(column_name, missing_count)
        
```

```python
# Remove duplicates from the data

df.drop(df[df.duplicated()].index,inplace=True)

```

```python
df
```

```python
'''replacing the duplicate values'''
df['Gender'].replace('Fe Male','Female',inplace=True)
df['Gender'].value_counts()
df['MaritalStatus'].replace('Single','Unmarried',inplace=True)
df['MaritalStatus'].value_counts()
```

```python
'''Finding exact number of value counts'''

catVar = ['TypeofContact','CityTier','Occupation','Gender','MaritalStatus','OwnCar','FeedbackScore','Designation']

for i in catVar:
    print()
    print(df[i].value_counts())
    print()
```
