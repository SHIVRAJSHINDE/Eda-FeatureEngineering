```python
'''Mapping external values to dataframe values in Pandas'''
'''Adding New Column'''
# Creating new dataframe
import pandas as pd
 
initial_data = {'First_name' : ['Ram', 'Mohan', 'Tina', 'Jeetu', 'Meera'],
                'Last_name'  : ['Kumar', 'Sharma', 'Ali', 'Gandhi', 'Kumari'],
                'Age'        : [42, 52, 36, 21, 23],
                'City'       : ['Mumbai', 'Noida', 'Pune', 'Delhi', 'Bihar']}
 
df = pd.DataFrame(initial_data, columns = ['First_name', 'Last_name','Age', 'City'])
 
# Create new column using dictionary
new_data = {"Ram"   :"B.Com",
            "Mohan" :"IAS",
            "Tina"  :"LLB",
            "Jeetu" :"B.Tech",
            "Meera" :"MBBS" }
 
df['Qualification'] = df["First_name"].map(new_data)


df
```

```python
'''Different ways to iterate over rows in Pandas Dataframe'''

data = {'Name': ['Ankit', 'Amit',
                 'Aishwarya', 'Priyanka'],
        'Age': [21, 19, 20, 18],
        'Stream': ['Math', 'Commerce',
                   'Arts', 'Biology'],
        'Percentage': [88, 92, 95, 70]}
 
# Convert the dictionary into DataFrame
df = pd.DataFrame(data, columns=['Name', 'Age', 
                                 'Stream', 'Percentage'])
df
```

```python
'''Method 1: Using the index attribute of the Dataframe.'''


for i in df.index:
    print(df['Name'][i],df['Stream'][i])
    
```

```python
'''Method 2: Using loc[],at[] function of the Dataframe. '''

for i in range(len(df)):
    print(df.loc[i,"Name"],df.loc[i,"Age"])

print()
print()

for i in range(len(df)):
    print(df.at[i,"Name"],df.at[i,"Age"])

    
```

```python
'''Method 3: Using iloc[] or iat[] function of the DataFrame. '''
for i in range(len(df)):
    print(df.iloc[i,0],df.iloc[i,1])

print()
print()

    
for i in range(len(df)):
    print(df.iat[i,0],df.iat[i,1])

```

```python
'''Method 6: Using apply() method of the Dataframe. '''
print(df.apply(lambda row: row["Name"] + " " +str(row["Percentage"]), axis=1))
```
