```python
'''Using map() function to create new column from existing column using a mapping function'''

data = {
    "name": ["John", "Ted", "Dev", "Brad", "Rex", "Smith", "Samuel", "David"],
    "salary": [10000, 20000, 50000, 45500, 19800, 95000, 5000, 50000]
}

df = pd.DataFrame(data)

def salary_stats(x):
    if x <= 45500:
        return "very low"
    elif x > 45500:
        return "very good"
    else:
        return "Average"


df['Comment'] = df['salary'].map(salary_stats)
df
```

```python
'''Using lambda function to create new column from existing column using a mapping function'''

data = {
    "name": ["John", "Ted", "Dev", "Brad", "Rex", "Smith", "Samuel", "David"],
    "salary": [10000, 20000, 50000, 45500, 19800, 95000, 5000, 50000]
}

df = pd.DataFrame(data)


df['Comment'] = df['salary'].apply(lambda x: "very low" if x <=45500 else "very good"  if  x >= 95000 else "Average")
df
```

```python
'''Creating a Pandas dataframe column based on a given condition using List Comprehension'''
# importing pandas as pd
import pandas as pd

data = {
    "name": ["John", "Ted", "Dev", "Brad", "Rex", "Smith", "Samuel", "David"],
    "salary": [10000, 20000, 50000, 45500, 19800, 95000, 5000, 50000]
}

df = pd.DataFrame(data)

df['Comment'] = ["LOW" if x <=4500 else "High" if x>= 95000 else "Average" for x in df['salary']]
df
```

```python
'''Getting frequency counts of a columns in Pandas DataFrame'''


# importing pandas as pd 
import pandas as pd 

# sample dataframe 
df = pd.DataFrame({'A': ['foo', 'bar', 'g2g', 'g2g', 'g2g', 
						'bar', 'bar', 'foo', 'bar'], 
				'B': ['a', 'b', 'a', 'b', 'b', 'b', 'a', 'a', 'b'] }) 

# frequency count of column A 
count = df['A'].value_counts() 
print(count) 
```
