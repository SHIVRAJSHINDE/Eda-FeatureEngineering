```python
count = df.groupby(['A', 'B']).size()
print(count) 
```

```python
'''Change Data Type for one or more columns in Pandas Dataframe'''

# importing pandas as pd
import pandas as pd

# sample dataframe
df = pd.DataFrame({
	'A': [1, 2, 3, 4, 5],
	'B': ['a', 'b', 'c', 'd', 'e'],
	'C': [1.1, '1.0', '1.3', 2, 5]})

print(df.dtypes)
df['A'] = df['A'].astype(str)
print(df.dtypes)
```

```python
'''Get the index of maximum value in DataFrame column'''
import pandas as pd

df = pd.DataFrame({
        'A': [1, 12, 3, 4, 5],
        'B': ['a', 'b', 'c', 'd', 'e'],
        'C': [1.1, 1.0, 1.3, 2, 5]})


print(df)
print()
print(df["A"].idxmax())
print(df['C'].idxmax())
```

```python
'''Get the index of maximum value in DataFrame column'''
import pandas as pd

df = pd.DataFrame({
        'A': [1, 12, 3, 4, 5],
        'B': ['a', 'b', 'c', 'd', 'e'],
        'C': [1.1, 1.0, 1.3, 2, 5]})

print(df['A'].idxmax())
print(df['C'].idxmax())

print(df['A'].idxmin())
print(df['C'].idxmin())
```
