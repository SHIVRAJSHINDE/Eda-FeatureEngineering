```python
# Date_of_Journey covert it in to 

final_df['Date'] = final_df['Date_of_Journey'].apply(lambda x: x.split("/")[0])
final_df['Month']= final_df['Date_of_Journey'].apply(lambda x:x.split("/")[1])
final_df['Year']= final_df['Date_of_Journey'].apply(lambda x:x.split("/")[2])

final_df['Date'] = final_df['Date'].astype(int)
final_df['Month']= final_df['Month'].astype(int)
final_df['Year']= final_df['Year'].astype(int)

final_df.drop('Date_of_Journey',axis=1,inplace=True)
final_df.head()
```

```python
final_df['Arrival_Hour'] = final_df['Arrival_Time'].apply(lambda x :x.split(' ')[0].split(':')[0])
final_df['Arrival_Minute'] = final_df['Arrival_Time'].apply(lambda x :x.split(' ')[0].split(':')[1])
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df.head(5)
```

```python
final_df['Dep_Hour'] = final_df['Dep_Time'].apply(lambda x :x.split(' ')[0].split(':')[0])
final_df['Dep_Minute'] = final_df['Dep_Time'].apply(lambda x:x.split(' ')[0].split(':')[1])

final_df['Dep_Hour'] = final_df['Dep_Hour'].astype(int)
final_df['Dep_Minute'] = final_df['Dep_Minute'].astype(int)


final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
def convertToMinute(valuesa):
    for i in valuesa:
        abc = i.split()
        if len(abc)==2:
            return int(i.split(' ')[0].replace("h",""))*60 +int(i.split(' ')[1].replace("m",""))
        elif len(abc)==1:
            return int(i.split(' ')[0].replace("h",""))*60
        
final_df.apply(convertToMinute(final_df['Duration']),axis=1)
```
