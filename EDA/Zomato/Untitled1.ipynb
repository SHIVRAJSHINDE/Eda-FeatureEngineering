```python
df_contry = pd.read_excel('Country-Code.xlsx')
```

```python
final_df = pd.merge(df,df_contry,on='Country Code',how='left')
final_df.head(3).T
```

```python
final_df.columns
```

```python


Country_Names = final_df.Country.value_counts().index
Country_Val = final_df.Country.value_counts().values

Country_Val
Country_Names
```

```python

## Pie Chart


plt.figure(figsize=(15,10))
plt.pie(Country_Val[:4],labels=Country_Names[:4],autopct="%1.f%%")



```

```python
plt.figure(figsize=(15,10))

sns.countplot(final_df['Country'])
```

```python
final_df.groupby(['Rating color','Rating text','Aggregate rating']).size()
```

```python

ratings = final_df.groupby(['Rating color','Rating text','Aggregate rating']).size().reset_index().rename(columns={0:'Rating Count'})

ratings
```

```python
## Observation
# 1) Rating between 4.5 to 4.9 Excellent
# 2) Rating between 4.0 to 3.4 Very Good
# 3) so on and soforth
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.countplot(x='Rating color',data=ratings)
```
