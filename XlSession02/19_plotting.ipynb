```python
# Playing the parameters
# matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center')
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
sizeOfFont = 15
plt.bar(df['Airline'],df['Price'],width=.5,align='center')
plt.xlabel("Airline",fontsize=sizeOfFont)
plt.ylabel('Price',fontsize=sizeOfFont)
plt.title("Airline Vs Price",fontsize=sizeOfFont)
plt.xticks(rotation=45)
plt.show()
```

```python
# change Color
import matplotlib.pyplot as plt
sizeOfFont = 15
plt.bar(df['Airline'],df['Price'],width=.5,align='center',color='c')
plt.xlabel("Airline",fontsize=sizeOfFont)
plt.ylabel('Price',fontsize=sizeOfFont)
plt.title("Airline Vs Price",fontsize=sizeOfFont)
plt.xticks(rotation=45)
plt.show()
```

```python
# Color
import matplotlib.pyplot as plt
#colourValue = ['y','b','g','r','y','b','g','r']
colourValue = ['y','b','g','r','y','b','g','r']

# change Color
import matplotlib.pyplot as plt
sizeOfFont = 15
plt.bar(df['Airline'],df['Price'],width=.5,align='center',color=colourValue)
plt.xlabel("Airline",fontsize=sizeOfFont)
plt.ylabel('Price',fontsize=sizeOfFont)
plt.title("Airline Vs Price",fontsize=sizeOfFont)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
